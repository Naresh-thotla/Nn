```python
import os
from bs4 import BeautifulSoup
import requests
import pprint
import re
import pyperclip
from urllib.request import urlopen, urlretrieve, quote

urls = ['http://www.censusindia.gov.in/2011census/C-series/C-14.html','http://www.censusindia.gov.in/2011census/C-series/C-13.html']
#scrape elements
msg = 'Enter Url, to exit type q and hit enter.'
url = input(msg)
while(url!='q'):
    OUTPUT_DIR = 'D:\mad'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

   #print titles only
    for link in soup.select('a[href^=""]'):
        sad = link.get('href')
        print (sad)
        if not any(sad.endswith(x) for x in ['.csv','.xls','.xlsx']):
            continue
        filename = os.path.join(OUTPUT_DIR, sad.rsplit('/', 1)[-1])
        print("Downloading %s to %s..." % ('http://www.censusindia.gov.in/2011census/C-series/'+sad, filename) )
        urlretrieve('http://www.censusindia.gov.in/2011census/C-series/'+sad, filename)
        print("Done.")
```
